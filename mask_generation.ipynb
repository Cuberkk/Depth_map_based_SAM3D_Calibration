```python
from PIL import Image
import numpy as np
```

```python
file_dir = "/home/erie_lab/Documents/kxz365/Github_repos/Depth_map_based_SAM3D_Calibration/meshes/chips_object/images"
```

```python
image_path = f"{file_dir}/image.png"
mask_path =  f"{file_dir}/masks_vis_01.png"
save_path = f"{file_dir}/0.png"
```

```python
img = Image.open(image_path).convert("RGB")
img_np = np.array(img)

mask = Image.open(mask_path).convert("L")
mask_np = np.array(mask) > 0
```

```python
object_img = np.zeros_like(img_np)    # H x W x 3

object_img[mask_np] = img_np[mask_np]

object_pil = Image.fromarray(object_img)
object_pil.save(save_path)
```
